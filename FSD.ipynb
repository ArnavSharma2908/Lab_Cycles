```python
<!DOCTYPE html>
<html>
<head>
  <title>Event Handling Demo</title>
  <style>
    .success { color: green; }
  </style>
</head>

<body onclick="bodyClicked()">
  <h2>JavaScript Event Handling Demo</h2>

  <!-- Button click event -->
  <button id="btn" onclick="changeText()">Click Me</button>
  <p id="msg" class="success"></p>

  <!-- Textbox input event -->
  <input type="text" id="nameInput" placeholder="Type something..." oninput="updateText()" />
  <p id="liveText"></p>

  <!-- Password input event -->
  <input type="password" id="pass" placeholder="Enter password" oninput="checkStrength()" />
  <p id="strength"></p>

<script>
  // 1. Body click event  shows alert
  function bodyClicked() {
    alert("You clicked on the page!");
  }

  // 2. Button click  change text + show success message
  function changeText() {
    const btn = document.getElementById("btn");
    btn.innerText = "Clicked!";
    document.getElementById("msg").innerText = " Button clicked successfully";
  }

  // 3. Update paragraph as user types
  function updateText() {
    const text = document.getElementById("nameInput").value;
    document.getElementById("liveText").innerText = "You typed: " + text;
  }

  // 4. Password strength checker
  function checkStrength() {
    const pass = document.getElementById("pass").value;
    const output = document.getElementById("strength");

    if (pass.length < 4) output.innerText = "Strength: Weak";
    else if (pass.length < 8) output.innerText = "Strength: Medium";
    else output.innerText = "Strength: Strong";
  }
</script>
</body>
</html>
```

#### 5. Create a JavaScript program where a random number between 1 and 100 is generated and the user is prompted to guess the number. Provide feedback on whether the guess is too high, too low, or correct, and allow the user to keep guessing until they get it right. After a correct guess, ask if they want to play again using confirm(); if yes, restart the game, otherwise, display "Thanks for playing!" and end the program.

```python
<!DOCTYPE html>
<html>
<head><title>Number Guessing Game</title></head>
<body>

<script>
do {
  let target = Math.floor(Math.random() * 100) + 1;
  let guess;

  while (guess != target) {
    guess = prompt("Guess a number between 1 and 100:");

    if (guess === null) break; // cancel pressed
    else if (guess < target) alert("Too low!");
    else if (guess > target) alert("Too high!");
    else alert(" Correct! You guessed it!");
  }

} while (confirm("Play again?"));

alert("Thanks for playing!");
</script>

</body>
</html>
```

#### 6. Write a JavaScript program to create an array of student objects, convert their marks to CGPA using the map method and display a list of students with a CGPA of 9 or higher. Also, show the total count of students with CGPA 9 and above using the filter and reduce methods.

```python
<!DOCTYPE html>
<html>
<head><title>CGPA Calculator</title></head>
<body>
<h3>Students with CGPA  9</h3>
<ul id="list"></ul>
<p id="count"></p>

<script>
const students = [
  { name: "Arnav", marks: 95 },
  { name: "Riya", marks: 85 },
  { name: "Kabir", marks: 92 },
  { name: "Meera", marks: 88 },
  { name: "Ananya", marks: 98 }
];

// 1. Convert marks to CGPA using map()
const result = students.map(s => ({
  ...s,
  cgpa: (s.marks / 10).toFixed(1)
}));

// 2. Get students with CGPA  9 using filter()
const toppers = result.filter(s => s.cgpa >= 9);

// 3. Count them using reduce()
const total = toppers.reduce(count => count + 1, 0);

// Display results
document.getElementById("list").innerHTML =
  toppers.map(s => "<li>" + s.name + " - CGPA: " + s.cgpa + "</li>").join("");

document.getElementById("count").innerText =
  "Total students with CGPA  9: " + total;
</script>
</body>
</html>
```

#### 9. Illustrate react program to implement error boundaries to find  
1. Division by zero error.  
2. Array index of out of bound. 

```python
// ErrorBoundary.js
import React, { Component } from "react";

class ErrorBoundary extends Component {
  constructor(props) {
    super(props);
    this.state = { hasError: false };
  }

  static getDerivedStateFromError(error) {
    return { hasError: true };
  }

  componentDidCatch(error, info) {
    console.log("Error caught:", error, info);
  }

  render() {
    return this.state.hasError
      ? <h3>Error occurred!</h3>
      : this.props.children;
  }
}

export default ErrorBoundary;
```

#### 24. Develop an Express.js API for Teacher Records that performs CRUD operations: 
i. Add routes to create, fetch all, fetch by ID, update, and delete teacher details.  
ii. Ensure JSON data is parsed using middleware and invalid requests return error messages.  
iii. Include fields: name, department, experience, and email.  
iv. Return meaningful success messages for each action.  

```python
const express = require('express');
const app = express();
app.use(express.json());

let teachers = [
  {id:1,name:"Alice",department:"CSE",experience:5,email:"alice@mail.com"},
  {id:2,name:"Bob",department:"ECE",experience:8,email:"bob@mail.com"},
  {id:3,name:"Charlie",department:"ME",experience:4,email:"charlie@mail.com"}
];

let nextId = 4;

const valid = ({name,department,experience,email}) =>
  name && department && experience && email;

// CREATE
app.post('/teachers',(req,res)=>{
  if(!valid(req.body))
    return res.status(400).json({message:"All fields required"});

  const t = { id: nextId++, ...req.body };
  teachers.push(t);
  return res.status(201).json({message:"Teacher created",data:t});
});

// READ ALL
app.get('/teachers',(req,res)=>{
  return res.json({message:"Teachers fetched",data:teachers});
});

// READ BY ID
app.get('/teachers/:id',(req,res)=>{
  const t = teachers.find(x=>x.id==req.params.id);
  if(!t)
    return res.status(404).json({message:"Teacher not found"});

  return res.json({message:"Teacher fetched",data:t});
});

// UPDATE
app.put('/teachers/:id',(req,res)=>{
  const t = teachers.find(x=>x.id==req.params.id);
  if(!t)
    return res.status(404).json({message:"Teacher not found"});

  if(!valid(req.body))
    return res.status(400).json({message:"All fields required"});

  Object.assign(t,req.body);
  return res.json({message:"Teacher updated",data:t});
});

// DELETE
app.delete('/teachers/:id',(req,res)=>{
  const i = teachers.findIndex(x=>x.id==req.params.id);
  if(i === -1)
    return res.status(404).json({message:"Teacher not found"});

  return res.json({
    message:"Teacher deleted",
    data: teachers.splice(i,1)[0]
  });
});

// INVALID ROUTE
app.use((req,res)=>{
  return res.status(404).json({message:"Invalid route"});
});

app.listen(3000,()=>console.log("Server running on port 3000"));
```

#### 25.Create a Course Management API using Node.js and MongoDB: 
 Add routes to create, view, update, and delete course details.  
 Each course contains: courseName, faculty, credits, and studentsEnrolled.  
 Implement a separate route /courses/totalStudents to calculate total enrollment across all courses.  
 Use Mongoose validation for required fields.  

```python
const express = require('express');
const mongoose = require('mongoose');
const app = express();
app.use(express.json());

// MongoDB connection
mongoose.connect('mongodb://127.0.0.1:27017/CourseDB');

// Schema
const courseSchema = new mongoose.Schema({
  courseName: { type:String, required:true },
  faculty: { type:String, required:true },
  credits: { type:Number, required:true },
  studentsEnrolled: { type:Number, required:true }
});

const Course = mongoose.model('Course', courseSchema);

// CREATE
app.post('/courses', async (req,res)=>{
  try{
    const c = await Course.create(req.body);
    res.status(201).json({message:"Course created",data:c});
  }catch(err){
    res.status(400).json({message:err.message});
  }
});

// READ ALL
app.get('/courses', async (req,res)=>{
  res.json(await Course.find());
});

// READ BY ID
app.get('/courses/:id', async (req,res)=>{
  const c = await Course.findById(req.params.id);
  if(!c) return res.status(404).json({message:"Course not found"});
  res.json(c);
});

// UPDATE
app.put('/courses/:id', async (req,res)=>{
  const c = await Course.findByIdAndUpdate(req.params.id, req.body, {new:true, runValidators:true});
  if(!c) return res.status(404).json({message:"Course not found"});
  res.json({message:"Course updated",data:c});
});

// DELETE
app.delete('/courses/:id', async (req,res)=>{
  const c = await Course.findByIdAndDelete(req.params.id);
  if(!c) return res.status(404).json({message:"Course not found"});
  res.json({message:"Course deleted",data:c});
});

// TOTAL STUDENTS
app.get('/courses/totalStudents', async (req,res)=>{
  const result = await Course.aggregate([
    {$group:{_id:null,total:{$sum:"$studentsEnrolled"}}}
  ]);
  res.json({totalStudents: result[0]?.total || 0});
});

app.listen(3000,()=>console.log("Server running on port 3000"));
```

#### 26.Develop a Node.js program using the MongoDB native driver to accomplish the following: 
1. Read a list of employee records from a file named employees.json.  
2. Connect to a MongoDB database CompanyDB and insert the records into a collection called employees.  
3. Update the record of the employee with empId = 101 by changing the value of the designation field.  
4. Retrieve all employee records and write them into a file named updated_employees.json.  
5. Display a message confirming that the export process has been successfully completed.  

```python
//employeeScript.js
const fs = require("fs");
const { MongoClient } = require("mongodb");

const url = "mongodb://localhost:27017";
const client = new MongoClient(url);

const inputFile = "employees.json";
const outputFile = "updated_employees.json";

// Step 1: Read file
const employees = JSON.parse(fs.readFileSync(inputFile));

async function run() {
  try {
    // Step 2: Connect to DB
    await client.connect();
    const db = client.db("CompanyDB");
    const col = db.collection("employees");

    // Insert records
    await col.insertMany(employees);

    // Step 3: Update empId = 101
    await col.updateOne(
      { empId: 101 },
      { $set: { designation: "Senior Developer" } }
    );

    // Step 4: Fetch all records
    const updated = await col.find().toArray();

    // Write to file
    fs.writeFileSync(outputFile, JSON.stringify(updated, null, 2));

    // Step 5: Success message
    console.log("Employee records exported successfully");

  } catch (err) {
    console.log("Error:", err);
  } finally {
    client.close();
  }
}

run();
```
