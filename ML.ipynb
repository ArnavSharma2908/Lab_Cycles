```python
'''For a given set of training data examples stored in a .CSV file, implement and demonstrate
the Find-S algorithm to output a description of the set of all hypotheses consistent with the
training examples.'''


import pandas as pd

data = pd.read_csv("data.csv")

df = pd.DataFrame(data)

X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values

hypothesis = None
for i in range(len(X)):
  if y[i] == "Yes":
    if hypothesis is None :
      hypothesis = X[i].copy()
    else :
      for j in range(len(hypothesis)):
        if hypothesis[j] != X[i][j]:
          hypothesis[j] = '?'
  print("Step ", i + 1, " Hypothesis: ", hypothesis)
print("\nFinal Hypothesis: ", hypothesis)
```

```python
"""
Lab 1 - BCSL607 Machine Learning
---------------------------------
Question:
    Label the first 50 points {x1, ..., x50} as follows:
    if (xi <= 0.5), then xi ∈ Class1, else xi ∈ Class2.
    Classify the remaining points using KNN (k=3).
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier


def run():
    """
    Generate 100 random points, label first 50 by threshold rule,
    classify remaining 50 using KNN (k=3), and visualize results.
    """
    # Step 1: Generate 100 random values between 0 and 1
    np.random.seed(42)
    X = np.random.rand(100).reshape(-1, 1)

    # Step 2: Assign labels for the first 50 points
    y = np.array(["Class 1" if x <= 0.5 else "Class 2" for x in X[:50]])

    # Step 3: Train the KNN classifier
    knn = KNeighborsClassifier(n_neighbors=3)
    knn.fit(X[:50], y)

    # Step 4: Predict labels for the remaining 50 points
    y_pred = knn.predict(X[50:])

    # Step 5: Combine labels for all 100 points
    final_labels = np.concatenate((y, y_pred))

    # Step 6: Visualize the classification
    plt.figure(figsize=(10, 4))
    plt.scatter(X[:50], [1] * 50, c='blue', label='Labeled Data (Class 1 & 2)')
    plt.scatter(X[50:], [2] * 50, c='red', label='Predicted Data')
    plt.xlabel("X values")
    plt.ylabel("Class")
    plt.legend()
    plt.title("KNN Classification of Randomly Generated Points")
    plt.tight_layout()
    plt.show()

    # Step 7: Print final classification
    for i in range(100):
        print(f"x{i+1}: {X[i][0]:.3f} → {final_labels[i]}")


if __name__ == "__main__":
    run()
```

```python
"""
Lab 2 - BCSL607 Machine Learning
---------------------------------
Question:
    Classify the remaining points x51, ..., x100 using KNN.
    Perform this for k = 1, 2, 3, 4, 5, 20, 30.
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier


def run():
    """
    Generate 100 random points, label first 50 by threshold rule,
    then classify remaining 50 using KNN for multiple values of k.
    Visualizes results for each k side by side.
    """
    # Step 1: Generate 100 random values between 0 and 1
    np.random.seed(42)
    X = np.random.rand(100).reshape(-1, 1)

    # Step 2: Assign labels for the first 50 points
    y = np.array(["Class 1" if x <= 0.5 else "Class 2" for x in X[:50]])

    # Step 3: Define k values to test
    k_values = [1, 2, 3, 4, 5, 20, 30]

    # Step 4: Train and classify for each k
    results = {}
    for k in k_values:
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X[:50], y)
        y_pred = knn.predict(X[50:])
        results[k] = y_pred

    # Step 5: Visualize classification for different k values
    plt.figure(figsize=(10, 6))
    colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink']
    for i, k in enumerate(k_values):
        plt.scatter(X[50:], [i] * 50, c=[colors[i]] * 50, label=f'k={k}')
    plt.xlabel("X values")
    plt.ylabel("Different k-values")
    plt.legend()
    plt.title("KNN Classification for Different k-values")
    plt.tight_layout()
    plt.show()

    # Step 6: Print results
    for k in k_values:
        print(f"\nClassification Results for k={k}:")
        for i in range(50):
            print(f"x{51 + i}: {X[50 + i][0]:.3f} → {results[k][i]}")


if __name__ == "__main__":
    run()
```

```python
"""
Lab 3 - BCSL607 Machine Learning
---------------------------------
Question:
    Develop a program to demonstrate the working of Linear Regression.
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


def run():
    """
    Generate synthetic linear data, split into train/test sets,
    fit a Linear Regression model, evaluate it, and visualize.
    """
    # Step 1: Generate Sample Data
    np.random.seed(42)
    X = 10 * np.random.rand(100, 1)
    Y = 2.5 * X + 5 + np.random.randn(100, 1)

    # Step 2: Split the dataset (80% Train, 20% Test)
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42
    )

    # Step 3: Train the Linear Regression Model
    model = LinearRegression()
    model.fit(X_train, Y_train)

    # Step 4: Make Predictions
    Y_pred = model.predict(X_test)

    # Step 5: Print Model Parameters
    print(f"Slope (m): {model.coef_[0][0]:.2f}")
    print(f"Intercept (b): {model.intercept_[0]:.2f}")
    print(f"Mean Squared Error: {mean_squared_error(Y_test, Y_pred):.2f}")

    # Step 6: Visualize the Results
    plt.figure(figsize=(8, 5))
    plt.scatter(X, Y, color='blue', label="Actual Data")
    plt.plot(X_test, Y_pred, color='red', linewidth=2, label="Regression Line")
    plt.xlabel("X (Independent Variable)")
    plt.ylabel("Y (Dependent Variable)")
    plt.legend()
    plt.title("Linear Regression Model")
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    run()
```

```python
"""
Lab 4 - BCSL607 Machine Learning
---------------------------------
Question:
    Develop a program to demonstrate the working of Polynomial Regression.
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


def run():
    """
    Generate synthetic quadratic data, apply polynomial feature transformation
    (degree=2), fit a Linear Regression model on transformed features,
    evaluate, and visualize the polynomial regression curve.
    """
    # Step 1: Generate Sample Data
    np.random.seed(42)
    X = 10 * np.random.rand(100, 1)
    Y = 2 + 3 * X + 4 * X**2 + np.random.randn(100, 1)

    # Step 2: Split the dataset (80% Train, 20% Test)
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42
    )

    # Step 3: Convert Features to Polynomial Features (Degree = 2)
    poly = PolynomialFeatures(degree=2)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    # Step 4: Train the Polynomial Regression Model
    model = LinearRegression()
    model.fit(X_train_poly, Y_train)

    # Step 5: Make Predictions
    Y_pred = model.predict(X_test_poly)

    # Step 6: Print Model Parameters
    print("Polynomial Regression Coefficients:", model.coef_)
    print("Intercept:", model.intercept_)
    print("Mean Squared Error:", mean_squared_error(Y_test, Y_pred))

    # Step 7: Visualizing the Results
    X_sorted = np.sort(X, axis=0)
    Y_pred_all = model.predict(poly.transform(X_sorted))

    plt.figure(figsize=(8, 5))
    plt.scatter(X, Y, color='blue', label="Actual Data")
    plt.plot(X_sorted, Y_pred_all, color='red', linewidth=2,
             label="Polynomial Regression Curve")
    plt.xlabel("X (Independent Variable)")
    plt.ylabel("Y (Dependent Variable)")
    plt.legend()
    plt.title("Polynomial Regression Model (Degree=2)")
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    run()
```

```python
"""
Lab 5 - BCSL607 Machine Learning
---------------------------------
Question:
    Develop a program to implement the Decision Tree algorithm.
    Use Breast Cancer Dataset for building the decision tree
    and apply this knowledge to classify a new sample.
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score


def run():
    """
    Load the Wisconsin Breast Cancer dataset, train a Decision Tree
    classifier (Gini, max_depth=4), evaluate accuracy, visualize
    the tree, and classify a new custom sample.
    """
    # Step 1: Load Breast Cancer Dataset
    data = load_breast_cancer()
    X = pd.DataFrame(data.data, columns=data.feature_names)
    Y = pd.DataFrame(data.target, columns=["Target"])

    # Step 2: Split Data (80% Train, 20% Test)
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42
    )

    # Step 3: Train the Decision Tree Classifier
    model = DecisionTreeClassifier(criterion="gini", max_depth=4, random_state=42)
    model.fit(X_train, Y_train)

    # Step 4: Make Predictions
    Y_pred = model.predict(X_test)

    # Step 5: Evaluate Model Accuracy
    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"Decision Tree Model Accuracy: {accuracy:.2f}")

    # Step 6: Visualize the Decision Tree
    plt.figure(figsize=(15, 10))
    plot_tree(
        model,
        feature_names=data.feature_names,
        class_names=["Malignant", "Benign"],
        filled=True
    )
    plt.title("Decision Tree Visualization (Breast Cancer Data)")
    plt.tight_layout()
    plt.show()

    # Step 7: Classify a New Sample
    new_sample = np.array([[
        20.0, 10.5, 135.0, 800.0, 0.1, 0.15, 0.2, 0.1, 0.2, 0.08,
        0.3, 1.2, 2.5, 30.0, 0.005, 0.02, 0.03, 0.01, 0.02, 0.004,
        15.0, 12.0, 100.0, 600.0, 0.13, 0.25, 0.3, 0.12, 0.2, 0.07
    ]])
    prediction = model.predict(new_sample)
    print("New Sample Classification:",
          "Benign" if prediction[0] == 1 else "Malignant")


if __name__ == "__main__":
    run()
```

```python
"""
Lab 7 - BCSL607 Machine Learning
---------------------------------
Question:
    Write a program to implement the Naive Bayesian classifier for a
    sample training dataset stored as a CSV file. Compute the accuracy
    of the classifier, considering few test datasets.

    Dataset: https://www.kaggle.com/datasets/himanshunakrani/naive-bayes-classificationdata
    Features: Glucose, Blood Pressure, Outcome (0 = No Diabetes, 1 = Diabetes)

Note:
    Download the dataset from the link above and save it as
    'diabetes_data.csv' in the same directory before running.
"""

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score


def run(csv_path: str = "diabetes_data.csv"):
    """
    Load the diabetes CSV dataset, train a Gaussian Naive Bayes classifier,
    evaluate accuracy on a test split, and classify a new sample.

    Args:
        csv_path (str): Path to the diabetes CSV file.
                        Defaults to 'diabetes_data.csv'.
    """
    # Step 1: Load dataset from CSV file
    df = pd.read_csv(csv_path)

    # Step 2: Split dataset into features (X) and target (Y)
    X = df.iloc[:, :-1]   # All columns except the last (features)
    Y = df.iloc[:, -1]    # Last column (target)

    # Step 3: Split data into Training (80%) and Testing (20%) sets
    X_train, X_test, Y_train, Y_test = train_test_split(
        X, Y, test_size=0.2, random_state=42
    )

    # Step 4: Train the Naïve Bayes Classifier
    model = GaussianNB()
    model.fit(X_train, Y_train)

    # Step 5: Make Predictions
    Y_pred = model.predict(X_test)

    # Step 6: Compute Accuracy
    accuracy = accuracy_score(Y_test, Y_pred)
    print(f"Naïve Bayes Classifier Accuracy: {accuracy:.2f}")

    # Step 7: Test with a New Sample (Glucose=130, BloodPressure=85)
    new_sample = np.array([[130, 85]])
    prediction = model.predict(new_sample)
    print(f"Predicted Class for New Sample: "
          f"{'Diabetes' if prediction[0] == 1 else 'No Diabetes'}")


if __name__ == "__main__":
    run()
```
