## Math

```python
import math
print(math.sqrt(25))
print(math.factorial(5))
print(math.pi)
print(math.e)
print(math.sin(math.radians(30)))
```

## Numpy

```python
import numpy as np
arr = np.array([1,2,3])
print(arr)

matrix = np.array([[1,2],[3,4]])
print(matrix)
print(np.zeros((2,3)))
print(np.ones((3,3)))
print(np.arange(0,10,2))


print(arr+np.array([5,8,3]))
print(np.mean(arr), np.std(arr), sep='\n')
```

## Pandas

```python
import pandas as pd
data = [6, 2, 8]
series = pd.Series(data, index = ['a', 'b', 'c'])
print(series)

data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25,30,35],
        'Salary' : [50000, 60000, 70000]}
df = pd.DataFrame(data)
print(df)
```

## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

# Line Chart
x = [1,2,3,4,5]
y = [10, 20, 25, 30, 50]
plt.plot(x,y,
         marker = '+', linestyle = '-',
         color = 'g', label = 'Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.legend()
plt.show()

# Bar Chart
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 25]
plt.bar(categories, values, color = 'green')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()

# Scatter Chart
x_data = np.random.rand(50)
y_data = np.random.rand(50)
plt.scatter(x,y)
plt.show()

# Pie Chart
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [30, 20, 35, 15]
plt.pie(sizes, labels=labels, colors = ['blue', 'red', 'green', 'yellow'])
plt.show()

# Histogram
data = np.random.randn(1000)
plt.hist(data, bins = 30, color = 'purple', edgecolor = 'black')
plt.show()
```
